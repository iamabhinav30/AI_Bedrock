## Common Streaming Events

| Event Type | Purpose |
|---|---|
| **messageStart** | A new message is being sent |
| **contentBlockStart** | Start of a new block, which contains a text message, tool use, or other |
| **contentBlockDelta** | Chunk related to the latest content block that was started |
| **contentBlockStop** | The current content block has been completed |
| **messageStop** | The current message is complete |
| **metadata** | Extra information about the request |

```python
messages= []

add_user_message(messages, "Write a 1 sentence decription of a fake database")
response = client.converse_stream(messages=messages, modelId=model_id_sonnet4)

text = ''
for event in response['stream']:
    if 'contentBlockDelta' in event:
        chunk = event['contentBlockDelta']['delta']['text']
        print(chunk, end='')
        text += chunk
print("\n\nFull response:", text)
```

```python
# Demonstrate all common streaming events
messages = []
add_user_message(messages, 'Write a 1 sentence description of a fake database')

response = client.converse_stream(modelId=model_id_sonnet4, messages=messages)

full_text = ''

for event in response['stream']:

    # A new message is being sent
    if 'messageStart' in event:
        print(f"[messageStart] role: {event['messageStart']['role']}")

    # Start of a new content block (text, tool use, etc.)
    elif 'contentBlockStart' in event:
        print(f"[contentBlockStart] index: {event['contentBlockStart']['contentBlockIndex']}")

    # Chunk of text from the latest content block
    elif 'contentBlockDelta' in event:
        delta = event['contentBlockDelta']['delta']
        if 'text' in delta:
            full_text += delta['text']
            print(delta['text'], end='', flush=True)

    # Current content block is complete
    elif 'contentBlockStop' in event:
        print(f"\n[contentBlockStop] index: {event['contentBlockStop']['contentBlockIndex']}")

    # Current message is fully complete
    elif 'messageStop' in event:
        print(f"[messageStop] stopReason: {event['messageStop']['stopReason']}")

    # Extra metadata about the request (token usage, latency)
    elif 'metadata' in event:
        usage = event['metadata'].get('usage', {})
        metrics = event['metadata'].get('metrics', {})
        print(f"[metadata] inputTokens={usage.get('inputTokens')} outputTokens={usage.get('outputTokens')} latencyMs={metrics.get('latencyMs')}")

print(f"\nFull response: {full_text}")
```
