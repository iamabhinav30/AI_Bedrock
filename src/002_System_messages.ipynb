```python
import sys, os
sys.path.append(os.path.dirname(os.path.abspath("__file__")))

from bedrock_client import client, model_id
```

```python
system_prompt = '''
You are a helpful assistant.
'''
```

```python
def add_user_message(messages, text):
    messages.append({
        'role': 'user',
        'content': [{'text': text}]
    })

def add_assistant_message(messages, text):
    messages.append({
        'role': 'assistant',
        'content': [{'text': text}]
    })

def chat(messages, system=None, temperature=1.0):
    params = {
        'modelId': model_id,
        'messages': messages,
        'inferenceConfig': {
            'temperature': temperature
        }
    }

    if system:
        params['system'] = [{'text': system}]

    response = client.converse(**params)

    return response['output']['message']['content'][0]['text']
```

```python
# Start a conversation
messages = []

# Add a user message
add_user_message(messages, 'Generate a movie idea in one sentence.')

# Get assistant response
text = chat(messages)

print(text)
```

```python
# Interactive chat loop
chat_messages = []

while True:
    user_input = input('> ')

    if user_input.strip().lower() == 'exit':
        print('Ending chat. Goodbye!')
        break

    print(f'> {user_input}')
    add_user_message(chat_messages, user_input)
    text = chat(chat_messages)
    add_assistant_message(chat_messages, text)
    print(text)
```
